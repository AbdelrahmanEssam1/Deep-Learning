# Sentiment Analysis

```python
from transformers import pipeline
cls = pipeline("sentiment-analysis")
cls("I love football")
```

```python
cls("I hate this resturant")
```

# Zero-Shot Classification

```python
clf = pipeline("zero-shot-classification")
clf(
    "I hate vegetables",
    candidate_labels = ['education' , 'food','business']
)
```

# Text Generation

```python
generator = pipeline("text-generation")
generator("i was going to my friend but i met my father in my way so")
```

```python
generator("i was about to score a goal but the refreee")
```

```python
generator = pipeline("text-generation" , model = 'distilgpt2')
generator(
    "in this training course we will teach you how to",
    max_length = 40,
    num_return_sequences = 2
)
```

# Fill-Mask

```python
masker = pipeline("fill-mask")
masker("in this day i will learn some basics about <mask>", top_k = 2)
```

# Named Entity Recognition

```python
ner = pipeline("ner" , grouped_entities = True)
ner("Cristiano Ronaldo is a Portuguese player plays in Sudia-Arabia for Nasr FC")
```

# Question Answering

```python
qa = pipeline("question-answering")
qa(
    question = "what is the capital of Egypt",
    context = "I like Pyramids where in Giza and i visted Cairo the capital of Egypt"
)
```
