#### Test on a video

```python
model = YOLO('/content/runs/detect/train3/weights/best.pt')

video_path = '/content/Cars, Busy Streets, City Traffic - No Copyright Royalty Free Stock Videos.mp4'

#Load video using cv

cap = cv2.VideoCapture(video_path)

#Get video details (width, height , frame per second)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object to save output

output_path = '/content/video.mp4'

fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(output_path , fourcc , fps , (width , height))

# Loop through video frames

while cap.isOpened():

  ret , frame = cap.read()

  if not ret:
    break #no more frame ==> The end of the video

  # Perform object detection on the frame
  results = model(frame)

  # Convert the detection results to a format OpenCV can display
  annotated_frame = results[0].plot() # Annotated frame with detections

  # Write the frame with detections to the output video
  out.write(annotated_frame)

  # Exit the loop if 'q' is pressed
  if cv2.waitKey(1) & 0xFF == ord('q'):
      break
# Release video objects
cap.release()
out.release()
cv2.destroyAllWindows()
```
