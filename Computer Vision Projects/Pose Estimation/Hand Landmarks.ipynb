```python
import mediapipe 
import cv2
import time
```

```python
result = cv2.VideoWriter("handlandmarks.avi" , cv2.VideoWriter_fourcc(*"XVID") ,30 , (640,480))
```

```python
cap = cv2.VideoCapture(0)

mphands = mediapipe.solutions.hands # Mediapipe hands module for detecting and tracking hands
hands = mphands.Hands(max_num_hands= 4) #nitializes the MediaPipe Hands solution, Class Hands provides some methods for detection and tracking hands landmarks 
mpDraw = mediapipe.solutions.drawing_utils #provides function to draw the detected landmarks and connections on the images

pTime = 0
cTime = 0

while True:
    _ , img = cap.read()
    
    image = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    
    results = hands.process(image)#detect hand landmarks.
    
    if results.multi_hand_landmarks:
        for handco in results.multi_hand_landmarks:
            for idd , co in enumerate(handco.landmark):
                
                h ,w ,c = img.shape
                
                cx , cy = int(co.x*w) , int(co.y*h)
                
                print(idd , cx , cy)
                
                if idd==0:
                    cv2.circle(img,(cx,cy),9 , (255,0,255), cv2.FILLED)
                    
                mpDraw.draw_landmarks(img , handco , mphands.HAND_CONNECTIONS)
                
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    
    pTime = cTime
    
    cv2.putText(img , str(int(fps)) , (15,60) , cv2.FONT_HERSHEY_COMPLEX , 3 , (255,0,0))
    
    result.write(img)
    cv2.imshow("Frame" , img)
    
    key = cv2.waitKey(1)
    if key == 27:
        break
        
    
cv2.destroyAllWindows()
cap.release()
result.release() 
```
